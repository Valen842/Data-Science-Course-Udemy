# LIBRARIES

```python
import numpy as np
import pandas as pd
import scipy
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
sns.set()
```

# DATA

```python
data = pd.read_csv('./real_estate_price_size.csv')
print(data)
```

```python
data.describe()
```

```python
y = data['size']
x1 = data['price']
plt.scatter(x1,y)
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
#OLS ordinary lease square regression
results.summary()
```

```python
plt.scatter(x1, y)
yhat = 0.0033 * x1 + -122.3349 #    b1 = 0.017    b0=0.275
fig = plt.plot(x1 , yhat, lw=4, c='orange', label='regression line')
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show()
```
